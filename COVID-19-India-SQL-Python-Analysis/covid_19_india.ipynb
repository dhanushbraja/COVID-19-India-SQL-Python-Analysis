```python
import pandas as pd

query = """
SELECT
`State/UnionTerritory`,
SUM(Cured) AS Total_Cured
FROM covid_19_india
GROUP BY `State/UnionTerritory`
ORDER BY Total_Cured DESC;
"""

df = pd.read_sql(query, conn)

print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.bar(df['State/UnionTerritory'], df['Total_Cured'])

plt.xticks(rotation=90)

plt.title("State-wise Total Cured")

plt.xlabel("State")

plt.ylabel("Total Cured")

plt.show()
```

```python
query = """
SELECT
`State/UnionTerritory`,
SUM(Deaths) AS Total_Deaths
FROM covid_19_india
GROUP BY `State/UnionTerritory`
ORDER BY Total_Deaths DESC;
"""

df = pd.read_sql(query, conn)
```

```python
plt.figure(figsize=(14,6))

plt.bar(df.iloc[:,0], df["Total_Deaths"])

plt.title("State-wise Total Deaths")
plt.xlabel("State")
plt.ylabel("Total Deaths")
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

query = """
SELECT
`State/UnionTerritory`,
SUM(Confirmed) AS Total_Confirmed
FROM covid_19_india
GROUP BY `State/UnionTerritory`
ORDER BY Total_Confirmed DESC
LIMIT 5;
"""

df = pd.read_sql(query, conn)

plt.figure(figsize=(8,8))

plt.pie(df["Total_Confirmed"],
        labels=df["State/UnionTerritory"],
        autopct="%1.1f%%",
        startangle=90)

plt.title("Top 5 States by Confirmed Cases")
plt.show()
```

```python
query = """
SELECT `State/UnionTerritory`,
SUM(Deaths) AS Total_Deaths
FROM covid_19_india
GROUP BY `State/UnionTerritory`
ORDER BY Total_Deaths DESC
LIMIT 10;
"""

df = pd.read_sql(query, conn)

plt.figure(figsize=(10,6))

plt.barh(df["State/UnionTerritory"], df["Total_Deaths"])

plt.title("Top 10 States by Deaths")
plt.xlabel("Total Deaths")
plt.ylabel("State")

plt.show()
```

```python
query = """
SELECT `State/UnionTerritory`,
SUM(Cured) AS Total_Cured
FROM covid_19_india
GROUP BY `State/UnionTerritory`
ORDER BY Total_Cured DESC
LIMIT 10;
"""

df = pd.read_sql(query, conn)

plt.figure(figsize=(10,6))

plt.plot(df["State/UnionTerritory"], df["Total_Cured"], marker="o")

plt.title("Top 10 States by Total Cured")
plt.xlabel("State")
plt.ylabel("Total Cured")

plt.xticks(rotation=90)

plt.show()
```

```python
query = """
SELECT Confirmed
FROM covid_19_india;
"""

df = pd.read_sql(query, conn)

plt.figure(figsize=(8,5))

plt.hist(df["Confirmed"], bins=20)

plt.title("Distribution of Confirmed Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Frequency")

plt.show()
```

```python
query = """
SELECT Confirmed, Deaths
FROM covid_19_india;
"""

df = pd.read_sql(query, conn)

plt.figure(figsize=(8,6))

plt.scatter(df["Confirmed"], df["Deaths"])

plt.title("Confirmed Cases vs Deaths")
plt.xlabel("Confirmed Cases")
plt.ylabel("Deaths")

plt.show()
```
